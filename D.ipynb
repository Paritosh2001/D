```python
from bs4 import BeautifulSoup

# HTML code with table tag and its contents
html = """
<html>
    <body>
        <table>
            <tr>
                <td>Row 1, Column 1</td>
                <td>Row 1, Column 2</td>
            </tr>
            <tr>
                <td>Row 2, Column 1</td>
                <td>Row 2, Column 2</td>
            </tr>
        </table>
        <p>Some other text</p>
    </body>
</html>
"""

# Create a Beautiful Soup object
soup = BeautifulSoup(html, 'html.parser')

# Find the table tag and remove it
table = soup.find('table')
table.extract()

# Print the updated HTML code without the table tag and its contents
print(soup)
```

```python
from bs4 import BeautifulSoup

# Open the HTML file in read mode
with open('C:/Users/parit/Downloads/ast_sci_data_tables_sample.html', 'r') as file:
    # Read the contents of the file
    html = file.read()

# Create a Beautiful Soup object
soup = BeautifulSoup(html, 'html.parser')
for table in soup.find_all('table'):
    table.extract()


# Extract all text content from the HTML code
text = soup.get_text()

# Save the text content to a text file
with open('C:/Users/parit/Downloads/ast_sci_data_tables_sample.html', 'w') as file:
    file.write(text)
    print(text)
```

```python
import pyttsx3

# Create an instance of the pyttsx3 engine
engine = pyttsx3.init()

# Set the voice
voices = engine.getProperty('voices')
engine.setProperty('voice', voices[1].id)  # Change the index to select a different voice

# Set the speaking rate to a slower value
rate = engine.getProperty('rate')
engine.setProperty('rate', rate-1)  # Change the value to set a different rate

# Convert text to speech
t = text
engine.say(t)
engine.runAndWait()


#import pyttsx3
#engine = pyttsx3.init()


#voices = engine.getProperty('voices')
#engine.setProperty('voice', voices[1].id)  
#t = text
#engine.say(t)
#engine.runAndWait()
```

```python
from googletrans import Translator
from gtts import gTTS
import os

translator = Translator()

# Translate text from English to Gujarati
translated_text = translator.translate(t, dest='gu').text

# Create a text-to-speech object and save the audio file
tts = gTTS(translated_text, lang='gu')
tts.save('hello10.mp3')

# Play the audio file
os.system('mpg123 hello.mp3')

```
